## sequence_data_part_1

- string -->immutable
- List   -->mutable
- array  -->mutable
- tuple  -->immutable
- dictionary --> mutable
- sets      --> mutable
- range   --> immutable

```python
strsample = "learning" #string
print(strsample) 
```

```python
lstnumbers = [1,2,3,3,3,4,5,6,7,8]        #list with duplicate values
print(lstnumbers)
```

```python
lstsample = ['same',1,2,4,5,4.7,True]    #list with different datatypes
print(lstsample)
```

```python
#in most programming languages array is known to store single datatypes
#one dimentional array
from array import *
arrsample = array('i',[1,2,3,4])         #creating array using array datatype where 'i' is specifying the datatype 
print(arrsample)                         # if one datatype is specified no other datatype can come 
for x in arrsample:
    print(x)              #printing the values of an array made using array function
```

```python
#tuple
tupsample = (1,2,4,5,6,'zaid',True)
print(tupsample)
```

```python
#tuple packing - where brackets are not used
tuplesample = 1,2,4,5,6,'zaid',True
print(tuplesample)
```

```python
#dictinary
dictsample = {
    1:"zaid",
    2:"raghav",
    3:"nitin"
}
print(dictsample)

'''So dictionary in Python is an unordered collection of data values,\n
which is used to store datavalues like a map.Unlike other data types that hold only single values in an element.\n
Dictionary holds key value pairs.So key values provided in the dictionary to make it more optimise.So in Python dictionary \n
can be created by placing sequence of elements within the curlybraces like this and you can separate your key value pairs by a comma and dictionaryholds a pair of values.\n
For example this is a pair of values one being the key and the other corresponding pair elementbeing its key value.\n
So here one is the key and key value here is first and values in a dictionary can beof any data type and can be duplicated as well.\n
But you can have only unique keys, but you can have duplicated values.'''
```

```python
#creating dictionary using built in function
dict_list = dict([(1, 'zaid'), (2, 'raghav'), (3, 'nitin')])
print(dict_list)
```

```python
#sets
setexample = {'zaid','hello world',1,3,4,1.2,3.14,True}      #list cannont be an element of a set
                                                             #and another set also cant be an element of a set
print(setexample)
```

```python
#creating a set using the set function
set("zaid")
```

```python
#range- built-in sequence type used for looping
rangesample = range(1,12,4)      #range(start, stop-1, step value)
print(rangesample)

for x in rangesample:         #printting the value of 'rangesample'
    print(x)
```

## sequence_data_part_2

sequence data operations-
- string indexing
- list indexing
- array indexing
- tupple indexing
- set indexing - not possible/ not supported
- dictionary indexing
- range indexing

```python
string = "learning"      #string
```

```python
string.index('g')       #finding index of a value in a string
```

```python
string.index('rning')
```

```python
list_indexing = [1,2,3,4,'zaid',True]
```

```python
print(list_indexing.index(2))
print(list_indexing.index('zaid'))
print(list_indexing.index(True))
```

```python
#indexing of array
from array import *
array_indexing = array('i',[1,2,3,4])      #array with integer type
for x in array_indexing:
    print(x)
```

```python
#tuple indexing
tuple = (1,2,3,4,5,'zaid')
print(tuple.index('zaid'))    #to know th index of the given value
print(tuple[2])
```

```python
#dictionary indexing - in this indexing does not happen from index placed but it happenes by keys of the dictionary
dict = {
    1:'zaid',
    'raghav':2,
    4:'nitin'
}

print(dict['raghav'])
print(dict[4])
```

```python
#range indexing
range_1 = range(1,12,4)

for x in range_1:
    print(x)
```

```python
range_1.index(5)
```

```python
range_1.index(12)         #index not inrange of the range
```
