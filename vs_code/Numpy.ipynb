```python
#Creation of array
my_list = [1,2,3,4,5,6]
print(my_list)
```

to create a numpy array we first need to import the numpy package

```python
import numpy as np
```

```python
#creating a numpy array
array = np.array(my_list, dtype = int)
print(array)
```

```python
print(type(array))
print(len(array))
print(array.ndim)
print(array.shape)     #as it is one dimentional array
```

but the shape of na array can be changes using reshape

```python
array2 = array.reshape(3,2)
print(array2)
print(array2.ndim)
print(len(array2))
print(array2.shape)
```

```python
array3 = array. reshape (3,-1)          # -1 represents that the the other dimention will be calculated by numpy itself
print (array3)
print (array3.ndim)
```

```python
##initializing numpy array from nested python lists

mylist_2 = [1,2,3,4,5]
mylist_3 = [2,3,4,5,6]
mylist_4 = [9,7,6,8,9]

nested_array = np.array ([mylist_2,mylist_3,mylist_4])
print(nested_array)
```

```python
print(nested_array.shape)
print(nested_array.ndim)
print(len(nested_array))
```

```python
nested_array2 = nested_array.reshape(1,15)       
print(nested_array2)
```

## Numpy Attributes

```python
# using range
r = range(24)
print(r)
```

```python
# array of evenly spaced numbers
a = np.arange(24)
print(a)
print(a.ndim)
```

```python
# reshaping array a
b = a.reshape(6,4,1)
print(b)
```

```python
c = b.reshape(1,24)
print(c)
```

## numpy arithmetic operations

```python
x = np.array([[1,2],[3,4]],dtype = np.float64)
y = np.array([[5,6],[7,8]],dtype = np.float64)
print(x)
print(y)
```

numpy.add()- performs elementwise addition between two arrays
Syntax: numpy.add(array_1,array_2)

```python
print(x+y)
print("next line")
binga = np.add(x,y)
print(binga)
```

numpy.subtract()- performs elementwise subtraction betwween two arrays

```python
print(np.subtract(x,y))
print("or")
print(np.subtract(y,x))
```

numpy.multiply()- performs elementwise multiplication between two arrays

```python
print(x)
print("    X")
print(y)
print("    ||")
print(np.multiply(x,y))
```
